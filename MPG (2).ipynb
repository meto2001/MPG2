# BIVARIATE

```python
fig, axs = plt.subplots(2, 2, sharey=True, tight_layout=True,figsize=(18,15))
sns.scatterplot(x="horsepower", y="mpg", data = Data, ax=axs[0,0])
sns.scatterplot(x="weight", y="mpg", data = Data, ax=axs[0,1])
sns.scatterplot(x="displacement", y="mpg", data = Data ,ax=axs[1,0])
sns.scatterplot(x="acceleration", y="mpg", data = Data ,ax=axs[1,1])
fig.suptitle('Affect of some columns on MPG',fontsize= 17)
axs[0,0].set_title("Horsepower",fontsize= 15)
axs[0,1].set_title("Weight",fontsize= 15)
axs[1,0].set_title("Displacement",fontsize= 15)
axs[1,1].set_title("Acceleration",fontsize= 15)
```

- Each of Horsepower,Weight and diplacement is negatively correlated to MPG.
- Acceleration doesn't seem to affect MPG.

```python
sns.catplot(x="cylinders",y="mpg",data = Data, kind= "box")
```

- Cars with 4 and 5 cylinders seem to have more MPG than others.
- Cars with 8 cylinders seem to have the least MPG.

```python
sns.catplot(x="model year",y="mpg",data = Data, kind= "bar",palette = 'Greys')
```

Cars with model year 80, 81, 82 have the most MPG which shows that newer cars have better MPG than older ones.

```python
sns.catplot(x="origin",y="mpg",data = Data, kind= "box")
```

Cars of origin 3 have the most MPG and  Cars of origin 2 are in the second place.
